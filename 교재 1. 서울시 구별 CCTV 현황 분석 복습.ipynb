# CCTV 데이터 파악하기
설치대 대수가 작은 5개 구는 어디인가? - 소계가 총 설치대수 이므로 소계를 기준으로 sort한다. 
정렬 함수는 sort_values(by='기준열이름', ascending=True) : 오름차순 정렬

```python
# cctv가 가장 작은 순
CCTV_Seoul.sort_values(by='소계', ascending=True).head(5)
```

```python
# cctv가 가장 많은 순
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

```python
# CCTV 최근 증갸율 계산해서 CCTV '최근 증가율' 칼럼 추가
# 데이터 프레임에 칼럼을 추가할 때는 df명['새로 추가될 컬럼명'] = 값들
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년'] + 
                            CCTV_Seoul['2015년'] + 
                            CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
pop_Seoul.head()
```

```python
# 0번 행의 합계 삭제
# 행 지우는 명령 : drop()
# 열 지우는 명령 : del()
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
# nan을 확인하는 함수 : df명[필드명], isnull()
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
# pop_Seoul의 26행에 nan값이 저장되어 있으므로 제거
pop_Seoul.drop([26], inplace=True)
pop_Seoul.head()
```

```python
# 인구수 대비 외국인과 고령자 비율 계산
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

```python
# 1. 어느구의 인구가 가장 많은가? 인구가 많은 5개 구 출력
pop_Seoul.sort_values(by='인구수', ascending=False).head()
```

```python
# 2. 외국인이 많은 5개 구를 출력
pop_Seoul.sort_values(by='외국인', ascending=False).head()
```

```python
# 3. 인구수 대비 외국인 많은 5개 구를 출력
pop_Seoul.sort_values(by='외국인비율', ascending=False).head()
```

```python
# 4. 고령자가 많은 5개 구를 출력
pop_Seoul.sort_values(by='고령자', ascending=False).head()
```

```python
# 5. 인구수 대비 고령자 많은 5개 구를 출력
pop_Seoul.sort_values(by='고령자비율', ascending=False).head()
```

```python
# CCTV_seoul과 pop_Seoul 병합코드
# 공통 컬럼인 '구별' 컬럼으로 조인
# 양쪽 모두 25개의 구로 이루어져 있다.
# merge() 함수
# pd.merge(조인할 df1, 조인할 df2, on='key로 사용할 공통 필드')
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
data_result.head()
```

```python
# 열 삭제 - del 명령어는 원본 반영됨
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

```python
# 그래프 그릴 때 축데이터를 쉽게 지정하기 위해서 행 인덱스를 구별로 변경
# 행 인덱스 지정 함수 : set_index('지정렬 이름', inplace=True)
data_result.set_index('구별', inplace=True)
data_result.head()
```

다수의 데이터 중 상관 관계가 가장 큰 데이터를 비교 인구와 관련된 각 필드와 CCTV 소계와의 상관 개수를 파악한 후 의미있는 그래프 표현

상관 계수 : 두 변수의 관련성을 확인 하는 방법 하나 절대값이 클수록 두 데이터는 관련이 있다고 본다 절대값을 기존으로 0.1이하 무시 0.1 ~ 0.3 약한 상관 관계 0.7 이상 강한 상관 관계

상관 관계 파악 변수(칼럼) CCTV총대수와 인구수 CCTV총대수와 고령자 비율 CCTV총대수와 외국인 비율

계산 함수 numpy 패키지 corrcoef(데이터1, 데이터2)

```python
# CCTV총대수와 고령자 비율
np.corrcoef(data_result['고령자비율'],data_result['소계'])
# 약한 음의 상관관계
```

```python
# CCTV총대수와 외국인 비율
np.corrcoef(data_result['외국인비율'],data_result['소계'])
# 약한 음의 상관관계
```

```python
# CCTV총대수와 인구수 상관관계 확인
np.corrcoef(data_result['인구수'],data_result['소계'])
# 약한 상관관계 - 그래프로 표현
```
