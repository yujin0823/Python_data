# pandas 고급 기능 - 두 DataFrame 병합하기

```python
df1= pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'],
                   'B' : ['B0', 'B1', 'B2', 'B3'],
                   'C' : ['C0', 'C1', 'C2', 'C3'],
                   'D' : ['D0', 'D1', 'D2', 'D3']},
                  index=[0, 1, 2, 3])

df2= pd.DataFrame({'A' : ['A4', 'A5', 'A6', 'A7'],
                   'B' : ['B4', 'B5', 'B6', 'B7'],
                   'C' : ['C4', 'C5', 'C6', 'C7'],
                   'D' : ['D4', 'D5', 'D6', 'D7']},
                  index=[4, 5, 6, 7])

df3= pd.DataFrame({'A' : ['A8', 'A9', 'A10', 'A11'],
                   'B' : ['B8', 'B9', 'B10', 'B11'],
                   'C' : ['C8', 'C9', 'C210', 'C11'],
                   'D' : ['D8', 'D9', 'D210', 'D11']},
                  index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 열 방향으로 단순히 합치는 것 : concat()
result = pd.concat([df1, df2, df3])
result
```

```python
# 단순히 아무 옵션 없이 사용하면 열 방향으로 병합
# key 지정된 구분은 다중 index가 되어 level을 형성한다.
result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({'B' : ['B2', 'B3', 'B6', 'B7'],
                    'D' : ['D2', 'D3', 'D6', 'D7'],
                    'F' : ['F2', 'F3', 'F6', 'F7']},
                  index=[2, 3, 6, 7])
result = pd.concat([df1, df4], axis=1)
```

```python
df1
```

```python
df4
```

```python
# concat명령어는 index를 기준으로 데이터를 합친다.
# 그래서 값을 가질 수 없는 곳은 Nan이 저장된다.
result
```

```python
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
result = pd.concat([df1, df4], ignore_index=True)
result
```

```python
left = pd.DataFrame({'key' : ['K0', 'K4', 'K2', 'K3'],
                    'A' : ['A0', 'A4', 'A2', 'A3'],
                    'B' : ['B0', 'B4', 'B2', 'B3']})

right = pd.DataFrame({'key' : ['K0', 'K4', 'K2', 'K3'],
                      'C' : ['C0', 'C4', 'C2', 'C3'],
                      'D' : ['D0', 'D4', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# key를 기준으로 merge명령에서 merge기준을 설정하는 on옵션으로 합치면 공통된 key에 대해서만 합치게 된다.
pd.merge(left, right, on='key')
```

```python
# 합치는 두 데이터를 하나씩 기준으로 합칠 수 있다, how : 한쪽 데이터를 설정
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

```python
pd.merge(left, right, how='inner', on='key')
```
